```python
import matplotlib.pyplot as plt

# Data for the bar chart
pruning_methods = ['SNIP-Global']*4 + ['SNIP-Semi-Global']*3 + ['None']*5
experiments = ['x-l', 'l-m', 'm-s', 's-n', 'l-m', 'm-s', 's-n', 'x', 'l', 'm', 's', 'n']
mAP_scores = [67.7, 65.6, 61.8, 48.9, 66.0, 61.4, 51.1, 68.5, 67.0, 63.5, 56.0, 45.5]

# Define colors for each pruning method using a dictionary
color_dict = {'SNIP-Global': 'red', 'SNIP-Semi-Global': 'green', 'None': 'blue'}
colors = [color_dict[method] for method in pruning_methods]

# Create a bar chart
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(range(len(mAP_scores)), mAP_scores, color=colors)

# Customize the chart
ax.set_xticks(range(len(mAP_scores)))
ax.set_xticklabels([f'{method}\n{exp}' for method, exp in zip(pruning_methods, experiments)])
ax.set_ylabel('mAP')
ax.set_title('YOLOv5 Model Pruning Results')

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the scatter plot
pruning_methods = ['SNIP-Global']*4 + ['SNIP-Semi-Global']*3 + ['None']*5
experiments = ['x-l', 'l-m', 'm-s', 's-n', 'l-m', 'm-s', 's-n', 'x', 'l', 'm', 's', 'n']
mAP_scores = [67.7, 65.6, 61.8, 48.9, 66.0, 61.4, 51.1, 68.5, 67.0, 63.5, 56.0, 45.5]
inference_times = [200.6782227, 126.8568066, 77.73362793, 45.73539551, 150.1195313, 90.8740918, 44.75087891, 210.114668, 128.3467871, 80.524, 42.18, 25.06]

# Define colors for each pruning method using a dictionary
color_dict = {'SNIP-Global': 'red', 'SNIP-Semi-Global': 'green', 'None': 'blue'}
colors = [color_dict[method] for method in pruning_methods]

# Create a scatter plot with swapped x and y axes
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(inference_times, mAP_scores, color=colors)

# Draw lines connecting points of the same pruning method
for method in color_dict.keys():
    method_indices = [i for i, x in enumerate(pruning_methods) if x == method]
    ax.plot([inference_times[i] for i in method_indices], [mAP_scores[i] for i in method_indices], color=color_dict[method], linewidth=2)

# Customize the chart
ax.set_xlabel('Inference Time (ms)')
ax.set_ylabel('mAP')
ax.set_title('YOLOv5 Model Pruning Results')
# ax.legend(color_dict.keys())
# ax.legend(['SNIP-Semi-Global'])
# Display the chart
plt.show()
```
